```python
# Here, we want to set the dimensions of the background _image_, so we should anticipate it beingan l x w rectangle.
b_width = 8.5
b_length = 11

# Switch variables if the length/width isn't properly defined.
if b_width > b_length:
    b_width, b_length = b_length, b_width
```

```python
# Pattern variables
num_iterations = None
color_palette = {'r','g','b'}
pattern_width = 0.5 # Variable space relative to b_width, b_length

## Design a "line options" dictionary here
offset = 0.5 # Distance between repeating lines, in inches
line_colors = {0:'k', 1:'k'}
```

```python
if num_iterations is None:
    # Check for pattern to reach the edge
    num_iterations = int(b_length/offset)

# Create a figure template using matplotlib
plt.figure(figsize = (b_length, b_width))

# Position is in landscape
plt.xlim((0,b_length))
plt.ylim((0,b_width))

# Creating a general geometric pattern
for i in enumerate(range(num_iterations)):
    mult = i[0]
    x = (offset*mult, offset*mult+1)
    y = (0,b_length)
    plt.plot(x,y, color = line_colors[0])
    plt.plot(x[::-1], y, color = line_colors[1])
```
